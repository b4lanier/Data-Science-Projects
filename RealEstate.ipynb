```python
import pandas as pd   #import pandas and set as variable pd
```

```python
data=pd.read_csv('Realestate.csv')  #point variable data to Realestate CSV data file
```

```python
age = data.houseAge   #set age variable to point to houseAge column of Realestate.csv
```

```python
norm=(age-age.mean())/age.std()      #normalize data for age by dividing variance from mean by st dev
data['houseAgeStandardized']=norm    #create new column for normalized houseAge data
```

```python
data.drop(columns=['numberOfConvenienceStores'], inplace=True)  #drop that column from the data set
```

```python
data.rename(columns={"transaction":"transactionDate"}, inplace=True)  #change name of column
```

```python
data.loc[0:10,:]  #display rows 0-10 using loc
```

```python
data.iloc[0:10,:]  #display first 10 rows using iloc
```

```python
#drop duplicates that have the same latitude, longitude, and transaction date, keeping first record
data.drop_duplicates(subset=['transactionDate','latitude','longitude'],keep='first', inplace=True)  
```

```python
#replace missing values in housePrice with the mean of the column
data.fillna(value={'housePriceOfUnitArea':data.housePriceOfUnitArea.mean()},inplace=True)
```

```python
#replace missing values in houseAge with the mean of the column
data.fillna(value={'houseAge':data.houseAge.mean()},inplace=True)
```

```python
#replace missing values in distanceToNearestMRTstation with the mean of the column
data.fillna(value={'distanceToNearestMRTstation':data.distanceToNearestMRTstation.mean()},inplace=True)
```

```python
#replace missing values in latitude with the mean of the column
data.fillna(value={'latitude':data.latitude.mean()},inplace=True)
```

```python
#replace missing values in longitude with the mean of the column
data.fillna(value={'longitude':data.longitude.mean()},inplace=True)
```
