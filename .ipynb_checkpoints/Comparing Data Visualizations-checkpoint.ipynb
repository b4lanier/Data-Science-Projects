```python
from matplotlib import pyplot as plt #note: pyplot is simple interface for matplotlib
plt.plot([1,2,3],[1,4,9])  #set up dummy data to show basics of graphs
plt.plot([1,2,3],[5,10,15])
plt.legend(['Data Set 1','Data Set 2'])
plt.xlabel('this is the x axis')
plt.ylabel('this is the y axis')
plt.title('this is the title')
plt.show()
```

```python
import pandas as pd
import numpy as np
data=pd.read_csv('countries.csv')  #Country dataset imported
data.head()
```

```python
afghanistan=data[data.country=='Afghanistan'] #set variable for all Afghanistan data
plt.plot(afghanistan.year,afghanistan.gdpPerCapita) #from variable data, x=year, y=GDP, plot graph
plt.title('Afghanistan GDP per Capita')
```

```python
set(data.continent) #see all data in continent column
```

```python
data_2007=data[data.year==2007] #set up variable for all 2007 data
asia_2007=data_2007[data_2007.continent=='Asia'] #set up variable for Asia data within 2007 dataset variable
europe_2007=data_2007[data_2007.continent=='Europe'] #set up variable for Europe data within 2007 dataset variable
asia_2007.head()
```

```python
print(len(set(europe_2007.country)))  #how many countries are in the europe 2007 dataset?
```

```python
print('Mean GDP per capita in Asia 2007:')
print(asia_2007.gdpPerCapita.mean())
print('Mean GDP per capita in Europe 2007:')
print(europe_2007.gdpPerCapita.mean())
```

```python
plt.subplot(2,1,1) #subplot of 2 rows, 1 column, select first row for Asia below
plt.title('GDP per Capita')
plt.hist(asia_2007.gdpPerCapita, 10, range=(0,50000), edgecolor='black')  #histogram of Asia 2007 GDP, 10 buckets, range, black border
plt.ylabel('Asia')
plt.subplot(2,1,2) #subplot of 2 rows, 1 column, select second row for Europe below
plt.hist(europe_2007.gdpPerCapita, 10, range=(0,50000), edgecolor='black')  #histogram of Europe 2007 GDP, 10 buckets, range, black border
plt.ylabel('Europe')

plt.show()
```

```python
data_1997=data[data.year==1997] #compare Europe and America life expectancy in 1997
americas_1997=data_1997[data_1997.continent=='Americas'] 
europe_1997=data_1997[data_1997.continent=='Europe']
bins=10
plt.subplot(2,1,1) #subplot of 2 rows, 1 column, select first row
plt.title('Life Expectancy')
plt.hist(americas_1997.lifeExpectancy, bins, range=(55,85), edgecolor='black')  #histogram of Asia 2007 GDP, 10 buckets, range, black border
plt.ylabel('Americas')
plt.subplot(2,1,2) #subplot of 2 rows, 1 column, select second row
plt.hist(europe_1997.lifeExpectancy, bins, range=(55,85), edgecolor='black')  #histogram of Europe 2007 GDP, 10 buckets, range, black border
plt.ylabel('Europe')
```

```python
us=data[data.country=='United States'] #Create variable with all US data
china=data[data.country=='China'] #Create variable with all China data
plt.title('GDP growth')
plt.plot(us.year,us.gdpPerCapita) #x-years, y-GDP
plt.plot(china.year,china.gdpPerCapita) #plot China on same graph
plt.legend(['US','China'])
plt.show()
```

```python
us_growth=us.gdpPerCapita/us.gdpPerCapita.iloc[0]*100 #normalize GDP to 100 by dividing by first value in the column
china_growth=china.gdpPerCapita/china.gdpPerCapita.iloc[0]*100
plt.title('GDP growth normalized')
plt.plot(us.year,us_growth) #x-years, y-GDP
plt.plot(china.year,china_growth) #plot China on same graph
plt.legend(['US','China'])
plt.show()
```

```python
plt.scatter(data_2007.gdpPerCapita, data_2007.lifeExpectancy) #scatterplot, x-GDP, y-Life expectancy
plt.title('GDP and life expectancy in 2007')
plt.show()
```

```python
data_2007.gdpPerCapita.corr(data_2007.lifeExpectancy) #are GDP and Life expectancy correlated in 2007?
```

```python
plt.scatter(np.log10(data_2007.gdpPerCapita), data_2007.lifeExpectancy) #scatterplot with log scale on x-axis to account for curve in graph
plt.show()
```

```python
years_sorted=sorted(set(data.year)) #variable with all values out of year column,  sorted in order
for given_year in years_sorted:   #loop for all years sorted
    data_year=data[data.year==given_year]  #set data year to current year loop
    plt.scatter(data_year.gdpPerCapita, data_year.lifeExpectancy,5)  #scatter plot current year's GDP/Life expectancy
    plt.title(given_year)
    plt.xlim(0,60000)
    plt.ylim(25,85)
    plt.savefig(str(given_year),dpi=200) #instead of print, save graph to string, dpi=dots/inch
    plt.clf()  #clear the current plot
```

```python
gdp=(data_2007.gdpPerCapita*data_2007.population)/10**9 #multiply to calc GAP, divide by 10^p to put in billions
plt.scatter(gdp,data_2007.lifeExpectancy,5)
```

```python
plt.scatter(np.log10(gdp),data_2007.lifeExpectancy,5) #scatter the graph above with a log scale
```

```python
top10=data_2007.sort_values('population', ascending=False).head(10) #create dataset for top 10 populations in 2007
plt.subplot(2,1,1) #compare 2 graphs, so this will stack 1st of 2 graphs
x=range(10)  #this is the same as setting x= [0,1,2...9]
plt.bar(x,top10.population/10**6)  #bar graph with the population values for the top10 in millions
plt.title('10 most Populous countries')
plt.xticks([],[]) #empty lists removes the x-axis labels
plt.legend(['Population in millions'])

plt.subplot(2,1,2) #2nd of 2 graphs
plt.bar(x,top10.gdpPerCapita*top10.population/10**9) #calculate total GDP in billions
plt.xticks(x,top10.country,rotation='vertical')  #x-labels, country name, vertical orientation
plt.legend(['GDP in Billions'])
plt.show()
```

```python
data_mean=data.groupby('year').mean()  #calculate mean for the dataset by year
plt.plot(data_mean.index, data_mean.population, 'red') #graph index(year) and population means, red line
plt.show()
```

```python
#Historgram to compare Asia and Africa life expectancy in 1952
data_1952=data[data.year==1952] #set up variable for all 1952 data
asia_1952=data_1952[data_1952.continent=='Asia'] #set up variable for Asia data within 1952 dataset 
africa_1952=data_1952[data_1952.continent=='Africa'] #set up variable for Africa data within 1952 dataset 
```

```python
bins=10
plt.subplot(2,1,1) #subplot of 2 rows, 1 column, select first row
plt.title('1952 Life Expectancy')
plt.hist(asia_1952.lifeExpectancy, bins, range=(20,70), edgecolor='black')  #histogram of Asia 1952 Life Expectancy, 10 buckets, range, black border
plt.ylabel('Asia')
plt.subplot(2,1,2) #subplot of 2 rows, 1 column, select second row
plt.hist(africa_1952.lifeExpectancy, bins, range=(20,70), edgecolor='black')  #histogram of Africa 1952 Life Expectancy, 10 buckets, range, black border
plt.ylabel('Africa')
```

```python
#time series to compare Asia and Africa life expectancy since 1952
asia=data[data.continent=='Asia'] #set variable for all Asia data
africa=data[data.continent=='Africa'] #set variable for all Africa data
asia_mean=asia.groupby('year').mean()  #calculate mean for the asia countries
africa_mean=africa.groupby('year').mean()  #calculate mean for the africa countries
asia_mean.head(20)
```

```python
plt.plot(asia_mean.index, asia_mean.lifeExpectancy) #from asia_mean data, x=year, y=LifeExpectancy, plot graph
plt.plot(africa_mean.index,africa_mean.lifeExpectancy) #add Africa_mean to the graph
plt.title('Mean Life Expectancy')
plt.legend(['Asia','Africa'])
plt.show()
```
