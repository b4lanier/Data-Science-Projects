```python
import pandas as pd  #import modules-Pandas, numpy, and seaborn?
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
diamonds=sns.load_dataset('diamonds')
diamonds.head()
```

```python
diamonds.describe() #summarize all main statistical measures for each column
```

```python
sns.set(color_codes=True)
sns.kdeplot(diamonds['price'],shade=True)  #KDE is a density function
plt.axvline(diamonds['price'].mean(),0,1,color='r')  #draw a line at the mean
```

```python
sns.displot(diamonds['price'], color='blue',bins=100)  #displot is based on actual values
```

```python
np.corrcoef(diamonds['carat'],diamonds['price'])  #see the correlation coefficient between carats and the price of diamonds
```

```python
diamonds.corr()  #see correlation of all columns
```

```python
sns.scatterplot(x='carat',y='price',data=diamonds)  #scatterplot of price and carat
```
